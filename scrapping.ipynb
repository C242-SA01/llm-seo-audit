```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL dari situs yang akan dianalisis
url = "https://www.grobmart.com"

# Mengambil konten HTML dari situs
try:
    response = requests.get(url)
    response.raise_for_status()  # Memeriksa apakah permintaan berhasil
    html_content = response.text
except requests.exceptions.RequestException as e:
    print(f"Terjadi kesalahan saat mengambil data dari URL: {e}")
    exit()
```

```python
# Mem-parsing konten HTML dengan BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# 1. Mengekstrak Meta Tags
def extract_meta_tags(soup):
    meta_data = {}
    # Meta description
    meta_description = soup.find("meta", attrs={"name": "description"})
    meta_data["description"] = meta_description["content"] if meta_description else "Tidak ditemukan"

    # Meta keywords
    meta_keywords = soup.find("meta", attrs={"name": "keywords"})
    meta_data["keywords"] = meta_keywords["content"] if meta_keywords else "Tidak ditemukan"

    #Meta Title
    meta_title = soup.find("meta", attrs={"name": "title"})
    meta_data["title"] = meta_title["content"] if meta_title else "Tidak ditemukan"

    #Meta Robots
    meta_robots = soup.find("meta", attrs={"name": "robots"})
    meta_data["robots"] = meta_robots["content"] if meta_robots else "Tidak ditemukan"

    #Open graph status
    meta_og_status = soup.find("meta", attrs={"property": "og:url"})
    meta_data["og_status"] = meta_og_status["content"] if meta_og_status else "Tidak ditemukan"

    #Canonical tag
    meta_canonical = soup.find("link", attrs={"rel": "canonical"})
    meta_data["canonical"] = meta_canonical["href"] if meta_canonical else "Tidak ditemukan"

    return meta_data

# 2. Mengekstrak Heading Tags
def extract_headings(soup):
    headings = {}
    for level in range(1, 7):  # H1 hingga H6
        tag = f"h{level}"
        headings[tag] = [heading.get_text().strip() for heading in soup.find_all(tag)]
    return headings

# 3. Menghitung Jumlah Kata
def count_words(soup):
    # Mengambil semua teks di dalam body konten
    body_text = soup.body.get_text() if soup.body else ""
    words = body_text.split()  # Membagi teks menjadi kata-kata
    return len(words)

# Mencari favicon
def find_favicon(soup):
    favicon_link = soup.find("link", rel="icon") or soup.find("link", rel="shortcut icon")
    if favicon_link:
        return favicon_link["href"]
    else:
        return "Tidak ditemukan"


# Menjalankan fungsi-fungsi untuk mendapatkan data SEO
meta_data = extract_meta_tags(soup)
headings = extract_headings(soup)
word_count = count_words(soup)
favicon = find_favicon(soup)

#menghitung jumlah nilai dalam tiap heading
for key, value in headings.items():
    headings[key]=len(value)

```

```python
#Mengecek sitemap.xml
def cek_sitemap(url):
  try:
    sitemap_url = url + "/sitemap.xml"
    response = requests.get(sitemap_url)
    response.raise_for_status()  # Memeriksa apakah permintaan berhasil
    print("Yes")
  except requests.exceptions.RequestException as e:
      print("No")
```

```python
#Mengecek robots.txt
def cek_robots(url):
  try:
    sitemap_url = url + "/robots.txt"
    response = requests.get(sitemap_url)
    response.raise_for_status()  # Memeriksa apakah permintaan berhasil
    print("Yes")
  except requests.exceptions.RequestException as e:
      print("No")
```

```python
#Google Search Console Connected
def cek_google_search_console(url):
  try:
    sitemap_url = url + "/google-search-console-connected"
    response = requests.get(sitemap_url)
    response.raise_for_status()  # Memeriksa apakah permintaan berhasil
    print("Yes")
  except requests.exceptions.RequestException as e:
      print("No")
```

```python
# Menampilkan hasil
print("Meta Tags:", meta_data)
print("\nHeading Tags:", headings)
print("\nJumlah Kata dalam Konten:", word_count)
print("\nMeta title :", meta_data["title"])
print("\nMeta Description: ", meta_data["description"])
print("\nMeta Keywords: ", meta_data["keywords"], "\nTotal keyword : ",len(meta_data["keywords"].split(",")))
print("\nMeta Robots: ", meta_data["robots"])
print("\nMeta Open Graph Status: ", meta_data["og_status"])
print("\nMeta Canonical: ", meta_data["canonical"])
print("\nSitemap: ")
cek_sitemap(url)
print("\nRobots.txt: ")
cek_robots(url)
print("\nGoogle Search Console Connected: ")
cek_google_search_console(url)
print("\nFavicon: ", favicon)
```
